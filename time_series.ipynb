```python
def rle_list(t_series):
    print(t_series)
    s = t_series.fillna('nan')
    mask = s.ne(s.shift())

    ids = s[mask].to_numpy()
    counts = s.groupby(mask.cumsum()).cumcount().add(1).groupby(mask.cumsum()).max().to_numpy()

    # Convert 'nan' string back to `NaN`
    ids[ids == 'nan'] = np.nan
    ser_out = pd.Series(counts, index=ids, name='counts').reset_index()
    print(ser_out)
    return np.where(ser_out["index"].isna(),-ser_out["counts"],ser_out["index"])
    # list = []
    # for index,value in ser_out.iteritems():
    #     if index == 'nan':
    #         list.append(int(-value))
    #     else:
    #         list.append(int(index.total_seconds()))
    # return list
```

```python
# rle_list(time_series)
def calculate_series(author,df_temp):
    time_series = pd.Series(np.nan,index=time_index).reset_index()

    # rows = df.loc[df["author_id"].isin([author])]
    # rows = df[df.author_id == author]
    # for _,row in df_temp.iterrows():
    #     time_series[row["created_at"]] = row["retweet_fuction"]

    time_series = time_series.merge(df_temp[["author_id","created_at","retweet_function"]],left_on="index",right_on="created_at",how="left").set_index("index")
    time_series = time_series["retweet_function"]
        
    time_list = rle_list(time_series)
    pd.DataFrame(time_list).to_csv("time_series/" + str(author) + ".csv",header=None,index=None) 
    return time_list
```

```python
from joblib import Parallel, delayed
from tqdm import tqdm
tqdm.pandas()
# input = tqdm(authors_id)
df_grouped = df_filtered.groupby("author_id")
# df_list = list(df.groupby("author_id"))
# df_grouped = df_list[:3]

series = Parallel(n_jobs=-1)(delayed(calculate_series)(author,df_temp) for author,df_temp in tqdm(df_grouped))
```

```python
import csv

with open("out.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(series)
```
